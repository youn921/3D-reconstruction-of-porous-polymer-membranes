# Synthetic data experiment-result-1

```python
import pandas as pd
result1_df = pd.read_csv('Synthetic Data 1-result_df.csv')
df_filtered1 = result1_df.iloc[:, 1:]
result1_df
```

```python
def highlight_rows(x):
    # Create a DataFrame with the same shape as x, filled with empty strings
    df_styler = pd.DataFrame('', index=x.index, columns=x.columns)
    # Apply the background color every 3 rows
    df_styler.iloc[::3, :] = 'background-color: #f0f0f0'  # Light gray color
    df_styler.iloc[1::3, :] = 'background-color: #e0e0e0'  # Slightly darker gray
    df_styler.iloc[2::3, :] = 'background-color: #d0d0d0'  # Even darker gray
    return df_styler

df_filtered1 = df_filtered1.style.apply(highlight_rows, axis=None)

display(df_filtered1)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Line
import pandas as pd

#Connecting quantile and sigma
sigma_quantile_combinations = result1_df[['Sigma', 'Quantile']].astype(str).agg(','.join, axis=1).unique()

result1_df['Time'] = result1_df['Time'].round(2)

# Get the running time of different acquisition functions
ucb_time = result1_df[result1_df['Acquisition Function'] == 'ucb'].groupby(['Sigma', 'Quantile'])['Time'].mean().tolist()
ei_time = result1_df[result1_df['Acquisition Function'] == 'ei'].groupby(['Sigma', 'Quantile'])['Time'].mean().tolist()
poi_time = result1_df[result1_df['Acquisition Function'] == 'poi'].groupby(['Sigma', 'Quantile'])['Time'].mean().tolist()

bar = (
    Bar()
    .add_xaxis(sigma_quantile_combinations.tolist())
    .add_yaxis('UCB', ucb_time, color='lightcoral',label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('EI', ei_time, color='lightblue',label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('POI', poi_time, color='lightgreen',label_opts=opts.LabelOpts(position="top"))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Execution Time of Synthetic Data Experiment(All Slices)"),
        xaxis_opts=opts.AxisOpts(name="Sigma,Quantile",name_location="middle",name_gap=30),
        yaxis_opts=opts.AxisOpts(name="Time (seconds)"),
        legend_opts=opts.LegendOpts(pos_right="5%"),
    )
)

# # 渲染图表
bar.render_notebook()
```

```python
result1_avg_time = result1_df.groupby('Acquisition Function')['Time'].mean().reset_index()
result1_avg_time.columns = ['Acquisition Function', 'Average Time']

result1_avg_time['Average Time'] = result1_avg_time['Average Time'].round(2)

acquisition_functions = result1_avg_time['Acquisition Function'].tolist()
average_times = result1_avg_time['Average Time'].tolist()

# Creating a Bar Chart
bar = (
    Bar()
    .add_xaxis(acquisition_functions)
    .add_yaxis(
        'Average Time',
        average_times,
        color='lightblue',
        label_opts=opts.LabelOpts(
            is_show=True, 
            
            position="top"
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Average Time by Acquisition Function of Synthetic Data Experiment(All Slices)"),
        xaxis_opts=opts.AxisOpts(name="Acquisition Function",name_location="middle",name_gap=30),
        yaxis_opts=opts.AxisOpts(name="Average Time (seconds)"),
        legend_opts=opts.LegendOpts(is_show=False)
    )
)

bar.render_notebook()
```

```python
def highlight_rows(x):
    # Create a DataFrame with the same shape as x, filled with empty strings
    df_styler = pd.DataFrame('', index=x.index, columns=x.columns)
    # Apply the background color every 3 rows
    df_styler.iloc[::3, :] = 'background-color: #f0f0f0'  # Light gray color
    df_styler.iloc[1::3, :] = 'background-color: #e0e0e0'  # Slightly darker gray
    df_styler.iloc[2::3, :] = 'background-color: #d0d0d0'  # Even darker gray
    return df_styler

df_filtered3 = df_filtered3.style.apply(highlight_rows, axis=None)

display(df_filtered3)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Line
import pandas as pd


result3_df['Real Image'] = ['Original SEM-1'] * 3 + ['Original SEM-2'] * 3 + ['Original SEM-3'] * 3 + ['Original SEM-4'] * 3
real_images = result3_df['Real Image'].unique().tolist()

result3_df['Time'] = result3_df['Time'].round(2)

# Get the running time of different acquisition functions
ucb_time3 = result3_df[result3_df['Acquisition Function'] == 'ucb'].groupby(['Real Image'])['Time'].mean().tolist()
ei_time3 = result3_df[result3_df['Acquisition Function'] == 'ei'].groupby(['Real Image'])['Time'].mean().tolist()
poi_time3 = result3_df[result3_df['Acquisition Function'] == 'poi'].groupby(['Real Image'])['Time'].mean().tolist()


bar = (
    Bar()
    .add_xaxis(real_images)
    .add_yaxis('UCB', ucb_time3, color='lightcoral',label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('EI', ei_time3, color='lightblue',label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('POI', poi_time3, color='lightgreen',label_opts=opts.LabelOpts(position="top"))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Time of Different Real SEM Images under Different Acquisition Functions"),
        xaxis_opts=opts.AxisOpts(name="The number of input original SEM image",name_location="middle",name_gap=30),
        yaxis_opts=opts.AxisOpts(name="Time (seconds)"),
        legend_opts=opts.LegendOpts(pos_right="5%"),
    )
)


bar.render_notebook()
```

```python
# Group 1 (first two images)，Group 2 (last two images)
group1_df = result3_df.iloc[0:6]  # Take the first six lines
group2_df = result3_df.iloc[6:12] # Take the last six lines


# Calculate the average time for each acquisition function in each group and keep two decimal places
group1_avg_time_ucb = group1_df[group1_df['Acquisition Function'] == 'ucb']['Time'].mean().round(2)
group1_avg_time_ei = group1_df[group1_df['Acquisition Function'] == 'ei']['Time'].mean().round(2)
group1_avg_time_poi = group1_df[group1_df['Acquisition Function'] == 'poi']['Time'].mean().round(2)

group2_avg_time_ucb = group2_df[group2_df['Acquisition Function'] == 'ucb']['Time'].mean().round(2)
group2_avg_time_ei = group2_df[group2_df['Acquisition Function'] == 'ei']['Time'].mean().round(2)
group2_avg_time_poi = group2_df[group2_df['Acquisition Function'] == 'poi']['Time'].mean().round(2)


bar = (
    Bar()
    .add_xaxis(['size:680×380', 'size:1348×1004'])  # X轴标签代表两组
    .add_yaxis('UCB', [group1_avg_time_ucb, group2_avg_time_ucb], color='lightcoral', label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('EI', [group1_avg_time_ei, group2_avg_time_ei], color='lightblue', label_opts=opts.LabelOpts(position="top"))
    .add_yaxis('POI', [group1_avg_time_poi, group2_avg_time_poi], color='lightgreen', label_opts=opts.LabelOpts(position="top"))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Average Execution Time by Different Size of Real SEM Image"),
        xaxis_opts=opts.AxisOpts(name="Two size of real SEM image", name_location="middle", name_gap=30),
        yaxis_opts=opts.AxisOpts(name="Average Time (seconds)"),
        legend_opts=opts.LegendOpts(pos_right="5%"),
    )
)


bar.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
import pandas as pd


# Calculate the average value of each error type under different acquisition functions
def calculate_means(df, error_type):
    return df.groupby('Acquisition Function')[error_type].mean().round(2).reindex(['ucb', 'ei', 'poi']).tolist()

error_types = ['3D Porosity Diff(%)', '2D Porosity Diff(10,%)', 'Specific Area Diff(10,%)']

x = ['ucb', 'ei', 'poi']

# Initialize an empty dictionary to store the average value of each error type
means = {}

# Iterate through each error type
for error_type in error_types:
    # Calculate the average value of each error type under different acquisition functions
    means[error_type] = calculate_means(evaluate1_df, error_type)


bar = (
    Bar()
    .add_xaxis(x)
    .add_yaxis('3D Porosity Diff(%)', means['3D Porosity Diff(%)'], color='lightcoral', label_opts=opts.LabelOpts(position="top",formatter="{c}%"))
    .add_yaxis('2D Porosity Diff(10,%)', means['2D Porosity Diff(10,%)'], color='lightblue', label_opts=opts.LabelOpts(position="top",formatter="{c}%"))
    .add_yaxis('Specific Area Diff(10,%)', means['Specific Area Diff(10,%)'], color='lightgreen', label_opts=opts.LabelOpts(position="top",formatter="{c}%"))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Validation of Synthetic Data(All Slices)"),
        xaxis_opts=opts.AxisOpts(name="Acquisition function", name_location="middle", name_gap=30),
        yaxis_opts=opts.AxisOpts(name="Average error"),
        legend_opts=opts.LegendOpts(pos_right="5%"),
    )
)


bar.render_notebook()
```
